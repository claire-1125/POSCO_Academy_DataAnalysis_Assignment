## 강수량에 따른 미세먼지 농도 비교 : 2 sample t-test

#### 1. 가설수립
- 귀무가설 : 비가 많이 온 날들의 미세먼지 농도 평균과 비가 적게 온 날들의 미세먼지 농도 평균이 동일하다.
- 대립가설 : 비가 많이 온 날들의 미세먼지 농도 평균과 비가 적게 온 날들의 미세먼지 농도 평균이 다르다.

#### 2. 유의수준 설정
유의수준(α)을 0.05로 설정했다.

#### 3. 가설검정 및 판단

```python
# 두 그룹을 각각 다른 DataFrame으로 분리 
df_lowTemp = df_Raining.loc[df_Raining.heavyRain == 0]
df_highTemp = df_Raining.loc[df_Raining.heavyRain == 1]

df_lowRain = df_lowTemp[["PM10"]]
df_highRain = df_highTemp[["PM10"]]

# 2 sample t-test
t_result = stats.ttest_ind(df_lowRain,df_highRain)

# 2 sample t-test의 t통계값, p-value 저장
t,p = t_result.statistic.round(3), t_result.pvalue.round(3)

print("< 강수량에 따른 2 sample t-test >")
print("t통계량: {}".format(t))
print("p-value: {}".format(p))
```

p-value가 유의수준보다 크므로 신뢰구간 안에 존재하는 상황이다. 이에 따라 귀무가설을 채택한다.**즉, 비가 많이 온 날들의 미세먼지 농도 평균과 비가 적게 온 날들의 미세먼지 농도 평균은 같다.**

## 풍속에 따른 미세먼지 농도 비교 : 2 sample t-test

#### 1. 가설수립
- 귀무가설 : 바람이 센 날들의 미세먼지 농도 평균과 약한 날들의 미세먼지 농도 평균이 동일하다.
- 대립가설 : 바람이 센 날들의 미세먼지 농도 평균과 약한 날들의 미세먼지 농도 평균이 다르다.

#### 2. 유의수준 설정
유의수준(α)을 0.05로 설정했다.

#### 3. 가설검정 및 판단

```python
# 두 그룹을 각각 다른 DataFrame으로 분리 
df_highTemp = df_WindNPM10.loc[df_WindNPM10.WIND >= df_WindNPM10.WIND.mean()]
df_lowTemp = df_WindNPM10.loc[df_WindNPM10.WIND < df_WindNPM10.WIND.mean()]

df_highWind = df_highTemp[["PM10"]]
df_lowWind = df_lowTemp[["PM10"]]

# 2 sample t-test
t_result = stats.ttest_ind(df_highWind,df_lowWind)

# 2 sample t-test의 t통계값, p-value 저장
t,p = t_result.statistic.round(3), t_result.pvalue.round(3)

print("< 풍속에 따른 2 sample t-test >")
print("t통계량: {}".format(t))
print("p-value: {}".format(p))
```

p-value가 유의수준보다 크므로 f-통계량이 신뢰구간 안에 존재하는 상황이다. 따라서 귀무가설을 채택한다. **즉, 바람이 센 날들의 미세먼지 농도 평균과 바람이 약한 날들의 미세먼지 농도 평균은 동일하다. 다시 말해 바람이 세게 불던 약하게 불던 미세먼지 농도와는 무관하다고 할 수 있다.**

- 대립가설 : 온도가 높은 날들의 미세먼지 농도 평균과 온도가 낮은 날들의 미세먼지 농도 평균이 다르다.

#### 2. 유의수준 설정
유의수준(α)을 0.05로 설정했다.

# 모델 예측 및 평가

```python
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def mean_absolute_percenatage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) *  100
```
